```python
import os
import shutil
import csv
```

```python
#Copy YOLO txt labels that match the benchmark list into a new directory
#might be useful later but no big deal, the crowd consensus labels are already in their own csv, forgot about it. 

def copy_images_from_list(source_dir, destination_dir, image_dir):
    # Create the destination directory if it doesn't exist
    os.makedirs(destination_dir, exist_ok=True)

    # Get the list of image filenames
    image_filenames = os.listdir(image_dir)

    for filename in image_filenames:
        base_filename = os.path.splitext(filename)[0]  # Remove the file extension
        label_file = os.path.join(source_dir, base_filename + '.txt')
        image_file = os.path.join(image_dir, filename)

        # Check if both label and image files exist
        if os.path.isfile(label_file) and os.path.isfile(image_file):
            # Copy the label file
            shutil.copy2(label_file, destination_dir)

            # Copy the image file
            #shutil.copy2(image_file, destination_dir)
            print(f"Successfully copied {base_filename}.txt")
        else:
            print(f"Files for {base_filename} not found!")

# Specify the source directory, destination directory, and the directory containing the image files
source_directory = 'C:\\Users\\rowanconverse\\Desktop\\data\\zooniverse\\'
destination_directory = 'C:\\Users\\rowanconverse\\Desktop\\data\\benchmark\\'
image_directory = 'C:\\Users\\rowanconverse\\Desktop\\data\\tiles\\'

# Call the function to copy the images
copy_images_from_list(source_directory, destination_directory, image_directory)
```

```python
#Translate YOLO to COCO and put back into a CSV

def convert_yolo_to_coco(label_dir, output_file, image_width, image_height):
    label_files = [f for f in os.listdir(label_dir) if f.endswith('.txt')]
    coco_data = []

    for label_file in label_files:
        with open(os.path.join(label_dir, label_file), 'r') as f:
            lines = f.readlines()

        for line in lines:
            class_id, x_center, y_center, bbox_width, bbox_height = map(float, line.strip().split())

            # Convert YOLO format to COCO format
            x_min = (x_center - bbox_width / 2) * image_width
            y_min = (y_center - bbox_height / 2) * image_height
            x_max = (x_center + bbox_width / 2) * image_width
            y_max = (y_center + bbox_height / 2) * image_height

            # Append the converted bounding box to the COCO data list
            coco_data.append([label_file, x_min, y_min, x_max, y_max, class_id])

    # Write the COCO data to a CSV file
    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['image_id', 'x_min', 'y_min', 'x_max', 'y_max', 'class_id'])
        writer.writerows(coco_data)

    print(f"Conversion completed. CSV file saved as {output_file}")

# Specify the label directory, output CSV file, and the original image dimensions
label_directory = 'C://Desktop//data//benchmark//'
output_csv_file = 'C://Desktop//data//expertconsensus_tiles.csv'
image_width = 684
image_height = 521

# Call the function to convert YOLO format to COCO format and save as CSV
convert_yolo_to_coco(label_directory, output_csv_file, image_width, image_height)
```
